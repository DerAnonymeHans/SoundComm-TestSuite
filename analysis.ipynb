## Read audio files

```python
import os
from scipy.io import wavfile

files = {}

for file in os.listdir("./input_audio_files"):
    if not file.endswith(".wav"):
        continue
    
    files[file] = wavfile.read(f"./input_audio_files/{file}")
```

```python
file = files["long.wav"]

sample_rate, aud_data = file

print(len(aud_data))
print(len(aud_data) / sample_rate / 60)
print(aud_data.shape[0])
```

```python
from helper import create_freq_buckets
import numpy as np

FREQ_RANGE = range(20, 1000)
FREQ_BUCKET_COUNT = 40

freq_buckets = create_freq_buckets(FREQ_RANGE.start, FREQ_RANGE.stop, FREQ_BUCKET_COUNT)
```

## Fourier Transform

```python
SEGMENT_LENGTH_IN_SEC = 10

len_data = len(aud_data)
channel_1 = np.zeros([2**(int(np.ceil(np.log2(len_data))))])
channel_1[:len_data] = aud_data[:, 0]

# length of each segment in the array
segment_length = sample_rate * SEGMENT_LENGTH_IN_SEC

for segment_start in range(0, len(channel_1), segment_length):
    segment = channel_1[segment_start:segment_start + segment_length]
    fourier = np.fft.fft(segment)    
    # remove imaginary part
    fourier = fourier[:len(fourier) // 2]
    
    for bucket in freq_buckets:
        amplitude_sum = sum(abs(fourier[freq]) for freq in bucket.freq_range)
        avg_amplitude = amplitude_sum / len(bucket.freq_range)
        bucket.add_amplitude(avg_amplitude)
        
for bucket in freq_buckets:
    print(bucket.freq_range, bucket.average())
```
